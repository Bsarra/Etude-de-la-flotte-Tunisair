```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

fname = r"airlineWebPage.html"
HtmlFile = open(fname, 'r', encoding='utf-8')
source_code = HtmlFile.read()
from bs4 import BeautifulSoup
soup = BeautifulSoup(source_code, 'html.parser')

table = soup.find('tbody')
rows = table.find_all('tr')

Airline = []
AircraftWithSeatmap = []
SeatPitch = []
SeatWidth = []
VideoType = []
LaptopPower = []
PowerType = []
WiFi = []
SeatType = []
rows
for tr in rows:
    cols = tr.find_all('td')
    
    Airline.append(cols[0].text.strip())
    AircraftWithSeatmap.append(cols[1].text.strip())
    SeatPitch.append(cols[2].text.strip())
    SeatWidth.append(cols[3].text.strip())
    VideoType.append(cols[4].text.strip())
    LaptopPower.append(cols[5].text.strip())
    PowerType.append(cols[6].text.strip())
    WiFi.append(cols[7].text.strip())
    SeatType.append(cols[8].text.strip())
    

       
    

FileTableOfAirLines = pd.DataFrame(data={'Airline':Airline,'Constructor':AircraftWithSeatmap,
                        'AircraftType':SeatPitch,'SeatWidth':SeatWidth,'VideoType':VideoType,
                        'laptopPower':LaptopPower,'PowerType':PowerType,'Wifi':WiFi,'SeatType':SeatType
                         })

FileTableOfAirLines=FileTableOfAirLines[['Airline','Constructor', 'AircraftType', 'SeatWidth','VideoType',
                                        'laptopPower','PowerType','Wifi','SeatType']]

FileTableOfAirLines
```
